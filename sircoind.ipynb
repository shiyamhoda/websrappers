Importing required libraries

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd


#url = 'https://www.sircoind.com:10035/products'

```

Getting list of part class 

```python
url = "https://www.syaccweb.com:10010/anon/class"
res_part_class = requests.request("GET", url)

#print(response.text)
data_part_class = res_part_class.json()
df_part_class = pd.json_normalize(data_part_class)
df_part_class
```

```python
df_part_class['entDataId']
```

```python
url = "https://www.syaccweb.com:10010/anon/part"

res = []

for  x in df_part_class['entDataId']:
    #p_class = f'00{x}'
    querystring = {"compId":"C4","partClass":x}

    response = requests.request("GET", url, params=querystring)
    #print(response)
    data = response.json()
    #print(data)
    for p in data['data']:
       res.append(p)

df= pd.json_normalize(res)

```

```python

df.dropna()
df.drop_duplicates()
#compId	partId	partClass	dtlDescription	saleDescription	weight	entStrData1	partImage	webPackQty	partCrosses	subClass	class3	qtyUom	priceUom	unitPrice

cols = ['compId', 'part_id', 'partClass', 'dtlDescription' ,'saleDescription','weight','entStrData1','partImage',
        'webPackQty',	'partCrosses', 'subClass' 'class3'	'qtyUom', 'priceUom', 'unitPrice']
drop_cols = ['saleDescription','entStrData1','partImage']
df = df.drop(columns= drop_cols)
df
```

```python
# Function to extract brand and part number
def extract_brand_part(cross):
    match = re.search(r'<u>(.*?)</u>: (.*?)(?=<|$)', cross)
    if match:
        return match.group(1), match.group(2)
    return None, None
```

```python
crosses_df = df['partCrosses'].str.split('<br>', expand=True)

# Create a new dataframe to store the results
result_df = pd.DataFrame()

# Iterate through each row and column of crosses_df
for i, row in crosses_df.iterrows():
    for j, cross in enumerate(row):
        if pd.notna(cross):
            brand, part = extract_brand_part(cross)
            if brand and part:
                new_row = df.loc[i].copy()
                new_row['CrossBrand'] = brand
                new_row['CrossPart'] = part
                result_df = pd.concat([result_df, new_row.to_frame().T], ignore_index=True)
```

```python
# Drop the original partCrosses column
result_df = result_df.drop('partCrosses', axis=1)
```

```python
from datetime import date 
cur_date = str(date.today())

file_name = f'sircoind_products-{cur_date}.csv'

file_name
```

```python
# Save the result to a new CSV file
result_df.to_csv(file_name, index=False)

print(f"Processing complete. Results saved to {file_name}")
```
